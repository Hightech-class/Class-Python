[문제] 장바구니 안의 물품 계산, 재고 및 매출 관리 프로그램 만들기
소규모 마트에서 사용할 수 있는 계산 프로그램을 만들어야 합니다. 마트에 있는 물품에 관한 **DB**, **장바구니에 있는 물품 계산**, **재고와 매출**을 확인할 수 있는 기능을 만들어 보세요

```python
물품_list = [
    "비누", "치약", "샴푸", "린스", "바디워시", "폼클렌징", "칫솔", "수건",
    "휴지", "물티슈", "세탁세제", "섬유유연제", "주방세제", "수세미", "고무장갑",
    "쌀", "라면", "햇반", "생수", "우유", "계란", "두부", "콩나물", "시금치",
    "양파", "감자", "고구마", "사과", "바나나", "오렌지", "귤", "토마토",
    "김치", "된장", "고추장", "간장", "식용유", "참기름", "소금", "설탕",
    "커피", "차", "과자", "빵", "젤리", "초콜릿", "음료수", "맥주", "소주",
    "고기(돼지고기)", "고기(소고기)", "닭고기", "생선", "오징어", "새우", "게",
    "쌀국수", "파스타", "잼", "버터", "치즈", "요거트", "아이스크림", "통조림",
    "냉동만두", "어묵", "햄", "소시지", "김", "미역", "다시마", "멸치",
    "밀가루", "부침가루", "튀김가루", "빵가루", "식초", "소스", "향신료",
    "양초", "성냥", "건전지", "전구", "쓰레기봉투", "지퍼백", "호일", "랩"
]
sales_log = [] # 가게 총 판매 내역(이름들)

print("장바구니계산_재고와 매출관리")

while True:
    cart = []
    print("\n계산을 시작합니다. (그만 담으려면 'q' 입력)")

    while True:
        pick_물품 = input("물건 이름: ")

        if pick_물품.lower() == 'q':
            break

        if pick_물품 in 물품_list:
            cart.append(pick_물품)
            print("장바구니에 해당 물품이 추가됐습니다.")
        else:
            print("재고가 없어서 장바구니에 물품을 추가하지 못했습니다.")

    if cart:
        print("계산을 완료했습니다.")
        sales_log.extend(cart)
    else:
        print("구매한 물품이 없습니다.")

    cont = input("다음 손님을 받으시겠습니까? (y/n): ")
    if cont.lower() == 'n':
        break

print("\n" + "="*30)
print("오늘의 매출")

# 1. 판매 개수 계산
total_count = len(sales_log)
print(f"오늘 총 판매 개수: {total_count}개")

# 2. 총 판매 금액 계산 (개당 1000원)
total_price = total_count * 1000

# [수정 포인트] sales_log 대신 계산된 결과인 total_price를 출력해야 금액이 나옵니다!
print("오늘 물품 총 판매액:", total_price, "원")
print(f"상세 판매 내역: {sales_log}") # 어떤 물건이 팔렸는지 보고 싶을 때 사용
print("영업을 종료합니다.")
```
